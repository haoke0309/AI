为了构建上面的神经网络，我们需要先编写一些工具函数，这些函数会对上面的数据进行特定功能的运算。最后再将这些单独的函数组合起来，构建出一个神经网络模型。

第一个编写的工具函数是sigmoid。前面文章我已经介绍过它了，sigmoid函数的作用就是把预测结果转换为0和1之间的值，不明白的复习下前面的文章。

## 编写工具函数

### 1.sigmoid

```python
def sigmoid(z):
    """
    功能 ：sigmoid激活函数
    
    参数 ：
    z -- 一个数值或者一个numpy数组
    
    返回值 ：
    s -- 经过sigmoid算法计算后的值，在[0，1]范围内
    """
    
    s = 1 / (1 + np.exp(-z))
    return s
```

### 2.初始化权重数组w和b

```python
def initialize_with_zeros(dim):
    """
    功能 ：初始化权重数组w和b
    
    参数 ：
    dim -- 维度 w的大小，本例中是12288
    
    返回值 ：
    w -- 权重数组
    b -- 偏置bias
    """
    
    w = np.zeros((dim,1))
    b = 0
    
    return w, b
```

### 3.前向传播，成本计算；反向传播，w和b偏导数。

```python
def propagate(w, b, X, Y):
    """
    功能 ：前向传播，成本计算；反向传播，w和b偏导数
    
    参数 ：
    w -- 权重数组，本例中的维度是（12288，1）
    b -- 偏置bias
    X -- 图片的特征数据，本例中维度是（12288，209）
    Y -- 图片的标签数据，本例中0无猫，1有猫，维度是（1，209）
    
    返回值 ：
    cost -- 损失
    grads -- dw和db组成的字典
    """
    
    m = X.shape[1]
    
    # 前向传播
    A = sigmoid(np.dot(w.T, X) + b)                             
    cost = -np.sum(Y*np.log(A) + (1-Y)*np.log(1-A)) / m  
    
    # 反向传播
    dZ = A - Y
    dw = np.dot(X,dZ.T) / m
    db = np.sum(dZ) / m
    
    # 将dw和db保存到字典里面
    grads = {"dw": dw,
             "db": db}
    
    return grads, cost
```

### 4.梯度下降

```python
def optimize(w, b, X, Y, num_iterations, learning_rate, print_cost = False):
    """
    功能 ：梯度下降
    
    参数 ：
    w -- 权重数组，本例中维度是（12288，1）
    b -- 偏置bias
    X -- 图片的特征数据，本例中维度是（12288，209）
    Y -- 图片的标签数据，0无猫，1有猫，维度是（1，209）
    num_iterations -- 指定优化次数
    learning_rate -- 指定学习率，用来控制梯度下降的参数
    print_cost -- 为True时，每优化100次就把成本cost打出来
    
    返回值 ：
    params -- 优化后的dw和db的数组
    costs -- 每优化100次，将成本记录下来，成本越小，表示参数越优化
    """
    
    costs = []
    
    for i in range(num_iterations):        
        grads, cost = propagate(w, b, X, Y) # 计算得出梯度和成本
                
        # 从字典中取出梯度
        dw = grads["dw"]
        db = grads["db"]
        
        # 进行梯度下降，更新参数，使其越来越优化，使成本越来越小
        w = w - learning_rate * dw
        b = b - learning_rate * db
        
        # 将成本记录下来
        if i % 100 == 0:
            costs.append(cost)
            if print_cost:
                print ("优化%i次后成本是: %f" %(i, cost))
    
    params = {"w": w,
              "b": b}
    return params, costs
```

### 5.预测

有了参数后，就可以使用它们来进行预测了，对任意图片进行预测是否有猫。

```python
def predict(w, b, X):
    """
    功能 ：预测
    
    参数 ：
    w -- 权重数组，本例中维度是（12288，1）
    b -- 偏置bias
    X -- 图片的特征数据，本例中维度是（12288，图片张数），即X可以是一张图片
         也可以是多张图片，当多张图片时，函数给出对每张图片的预测结果
         
    返回值 ：
    Y_prediction -- 对每张图片的预测结果
    """
    
    m = X.shape[1]
    Y_prediction = np.zeros((1,m))
       
    
    #预测
    A = sigmoid(np.dot(w.T, X) + b)
    
    #将上面sigmoid结果进一步转换，大于0.5输出1表示有猫，小于0.5输出0表示无猫
    for i in range(A.shape[1]):
        if A[0,i] >= 0.5:
            Y_prediction[0,i] = 1
    
    return Y_prediction
```

到此，我们已经编写了所需的所有工具函数了

## 建模及训练

```python
def model(X_train, Y_train, X_test, Y_test,num_iterations = 2000,learning_rate = 0.5, print_cost = False):
    """
    功能 ： 建立训练模型
    
    参数 ：
    X_train -- 训练图片，本例中维度是（12288，209）
    Y_train -- 训练图片对应的标签，本例中维度是（1，209）
    X_test -- 测试图片，本例中维度是（12288，50）
    Y_test -- 测试图片对应的标签，本例中维度是（1，50）
    num_iterations -- 需要训练的次数
    learing_rate -- 学习率，控制梯度下降的速度
    print_cost -- 为True时，每优化100次输出cost，以便观察成本的变化
    
    返回值 ：
    d -- 由：costs，Y_prediction_train, Y_prediction_test, 
            w, b, learning_rate, num_iterations 组成的字典
    """
    
    # 初始化待训练的参数
    w, b = initialize_with_zeros(X_train.shape[0])

    # 使用训练数据来训练/优化参数
    parameters, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate, print_cost)
    
    # 从字典中分别取出训练好的w和b
    w = parameters["w"]
    b = parameters["b"]
    
    # 使用训练好的w和b来分别对训练图片和测试图片进行预测
    Y_prediction_train = predict(w, b, X_train)
    Y_prediction_test = predict(w, b, X_test)
    
    # 打印出预测的准确率
    print("对训练图片的预测准确率为: {}%".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("对测试图片的预测准确率为: {}%".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    
    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
```

```python
d = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 2000, learning_rate = 0.005, print_cost = True)
```

### 学习率的选择

```python
learning_rates = [0.01, 0.001, 0.0001]
models = {}
for i in learning_rates:
    print('学习率为 ： '  + str(i) + '时')
    models[str(i)] = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 2000, learning_rate = i, print_cost = False)
    print ('\n' + "-------------------------------------------------------" + '\n')
for i in learning_rates:
    plt.plot(np.squeeze(models[str(i)]['costs']),label = str(models[str(i)]['learning_rate']))

plt.ylabel('cost')
plt.xlabel('iterations (hundres)')

#图例
legend = plt.legend(loc = 'upper center', shadow = True)

frame = legend.get_frame()
frame.set_facecolor('0.90')

plt.show()
```
